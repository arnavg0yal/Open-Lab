```python
import numpy as np
import matplotlib.pyplot as plt
print(14.1 * 1e6)
```

```python
In115_14MeV_cross_section = 4.625486 # barns

flux = 3e8 #n/s
target_area = 1.0 # cm^2
flux = flux / target_area # n/cm^2/s
target_thickness = 0.01 # cm
indium_density = 7.31 # g/cm^3
indium_molar_mass = 114.818 # g/mol
avogadro_number = 6.022e23 # atoms/mol

number_density = (indium_density / indium_molar_mass) * avogadro_number # atoms/cm^3
reaction_rate = flux * number_density * In115_14MeV_cross_section * 1e-24 # reactions/cm^3/s
total_reactions_per_second = reaction_rate * target_area * target_thickness
print(f"Total reactions per second: {total_reactions_per_second:.2e} reactions/s")

time_of_irradiation = 10 # hours
time_of_irradiation_seconds = time_of_irradiation * 3600 # seconds
total_reactions = total_reactions_per_second * time_of_irradiation_seconds
print(f"Total reactions after {time_of_irradiation} hours: {total_reactions:.2e} reactions")
```

```python
half_life = 54.14 #minutes
half_life_seconds = half_life * 60 # seconds
decay_constant = np.log(2) / half_life_seconds
activity = total_reactions * decay_constant
print(f"Activity after irradiation: {activity:.2e} Bq")
print(f"Activity after irradiation: {activity/3.7e10:.2e} Ci")
print(f"Activity after irradiation: {activity/3.7e4:.2e} uCi")
```
